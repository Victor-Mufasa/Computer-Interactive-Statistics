### Creating a vector
To create a vector we use c(...) e.g. age <- c(4,5,6,7,...)

Consider the following weight of patients in a hospital: 48, 65, 97, 55, 39, 78, 80
Write a program in R that;
* Creates a vector called weight
* Calculate the mean weight using the formula sum of X divided by n
* Calculate the standard deviation

```r
weight <- c(48, 65, 97, 55, 39, 78, 80)
mean_weight = mean(weight)
std <- sqrt(sum((weight - mean_weight)^2)/ (length(weight) - 1))
print(std)
```

```r
hist(weight, col = blues9, xlab = 'weights', ylab = 'frequency')
```

### Matrices
Matrices are two dimensional arrays of numbers with rows and columns. It can be created in two ways;
* dim()
* matrix()

```r
X <- 1:12
dim(X) <- c(3, 4)
X
```

```r
A <- matrix(data = 1:12, nrow = 3, ncol = 4)
A
```

```r
# We could also specify the order in which they will be filled either row-wise or column-wise
A <- matrix(data = 1:12, nrow = 3, ncol = 4, byrow = T)
A
```

```r
# Useful functions in matrices
print(nrow(A))
print(ncol(A))
print(t(A))
print(rbind(A, c(13, 14, 15, 16)))
print(cbind(A, c(13, 14, 15)))
```

### When using rbind() the matrices must have the same number of columns and also when using cbind() the matrices should have the same number of rows.

```r
# Matrix multiplication. We use %*% to multiply matrices.
b <- c(2, 6, 3, 6)
A%*%b
```

```r
# Solving matrices
m <- matrix(data = c(2, 4, 3, 5), nrow = 2, ncol = 2, byrow = TRUE)
b <- c(14, 19)
solve(m,b)
```

```r
# Accessing elements of a matrix
m[2,2]
# Modifying a matrix
m[2,2] <- 6
m
```

```r
## Reading in data
data <- read.csv('heightWeight.csv')
head(data)
```

```r
lin <- lm(data$weight~data$height)
summary(lin)
```

```r
library(tidyverse)
data %>%
    select(height, weight) %>%
    filter(weight > 140)
```

```r
summary(data)
```

```r
attach(data)
```

```r
mean(height, na.rm=FALSE)
```

```r
library(dplyr, help, pos = 2, lib.loc = NULL)
data %>%
    mutate(BMI = weight / (height)^2) %>%
    summarise(avg_bmi=mean(BMI, na.rm=FALSE))
```

```r
library(ggplot2)
data %>%
    ggplot(aes(x = height,y = weight)) + 
    geom_point()+geom_smooth(method = lm)
```

```r
library(gapminder)
```

```r
head(gapminder)
```

```r
summary(gapminder)
```

```r
mean(gapminder$gdpPercap)
```

```r
hist(gapminder$lifeExp, xlab = 'lifeExp', ylab = 'Frequency', main = 'Histogram of life expectancy')
```

```r
hist(log(gapminder$pop))
```

```r
boxplot(gapminder$lifeExp ~ gapminder$continent)
```

```r
plot(log(gapminder$gdpPercap) ~ gapminder$lifeExp)
```

```r
gapminder %>%
    select(country, lifeExp) %>%
    filter(country == 'South Africa' |
           country == 'Ireland') %>%
    group_by(country) %>%
    summarise(average_lifeExp = mean(lifeExp))
```

```r
# T-test to compare the similarity between the life expactancies of South Africa and Ireland
df <- gapminder %>%
        select(country, lifeExp) %>%
        filter(country == 'South Africa' |
               country == 'Ireland')
t.test(data = df, lifeExp ~ country)
```

```r
gapminder %>%
    filter(gdpPercap < 50000) %>%
    ggplot(aes(x = log(gdpPercap), y = lifeExp)) +
    geom_point(alpha=0.5, aes(col =year)) + 
    geom_smooth(method = lm) +
    facet_wrap(~continent, ncol = 2)
```
